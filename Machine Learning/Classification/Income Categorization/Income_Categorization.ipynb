### Statistical Analysis

```python
# Basic stats for all columns
dataset.describe(include='all').transpose()
```

 * For numerical data, we can see the mean, median, std, min, max.
 * For categorical data, we can see the frequency of the most occuring value and no of unique values. 

```python
# Basic stats for just numerical data
dataset.describe().transpose()
```

```python
# Getting count of each unique value for categorical data
dataset.groupby(['Education']).size()
```

```python
# Below code is an attempt to reduce the time it takes to calculate extra stats like mode, variance, skewness, kurtosis
# as well as count analysis for categorical data by operating on them at once

# Separate stats for each column, where categorical and numerical both have their stats
for x in dataset.columns[:1]:
    print(f"Basic stats for column "+x)
    if dataset[x].dtype == 'int64':
        col_mean = dataset[x].mean()
        print("Mean:", col_mean)
        col_median = dataset[x].median()
        print("Median:", col_median)
        col_mode = dataset[x].mode()
        print(f"Mode:\n{col_mode}")
        col_var = dataset[x].var()
        print("Variance:", col_var)
        col_std = dataset[x].std()
        print("Standard Deviation:", col_std)
        col_min = dataset[x].min()
        print("Min:", col_min)
        col_max = dataset[x].max()
        print("Max:", col_max)
        col_range = col_max - col_min
        print("Range:", col_range)
        col_q1 = dataset[x].quantile(0.25)
        print("Q1 (25%):", col_q1)
        col_q3 = dataset[x].quantile(0.75)
        print("Q3 (75%):", col_q3)
        col_iqr = col_q3 - col_q1
        print("IQR:", col_iqr)
        col_skew = dataset[x].skew()
        print("Skewness:", col_skew)
        col_kurt = dataset[x].kurt()
        print("Kurtosis:", col_kurt)
        print('\n')
    elif dataset[x].dtype == 'object':
        col_unique = dataset.groupby(x).size()
        print("Unique Values:", col_unique)
        print('\n')
    
```

### Correlation heatmap

```python
# Checking correlation between independent variables
def correlation_heatmap(train):
    correlations = train.corr()

    fig, ax = plt.subplots(figsize=(10,10))
    sns.heatmap(correlations, vmax=1.0, center=0, fmt='.2f',
                square=True, linewidths=.5, annot=True, cbar_kws={"shrink": .70})
    plt.show()
    
correlation_heatmap(dataset)
```

We find no correlation between independent variables

Since the Native Country for most of the input data is United States, it's not a good measure for our Income classification so we'll remove the column

```python
del dataset['Native_Country']
dataset.head()
```

```python
# Check unique values for each column in dataset
def return_unique(data):
    data_unique = {}
    for col in data.columns:
        unique_vals = dataset[col].unique()
        unique_vals = np.sort([str(i).strip() for i in unique_vals])
        data_unique[col] = unique_vals
    return data_unique

unique_values = return_unique(dataset)
unique_values
```

### Label Encoding

```python
from sklearn import preprocessing
```

```python
# Encoding all the categorical columns using LabelEncoder
for col in dataset.columns[:-1]:
    if dataset[col].dtype == 'object':
        dataset[col] = preprocessing.LabelEncoder().fit_transform(dataset[col])
```

```python
X = dataset.drop('Income', 1)
y = dataset['Income']
```

```python
y
```

### Normalizing Data

```python
dataset_norm = (X - X.min()) / (X.max() - X.min())
dataset_norm
```

### Splitting Data

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) 
```

```python
X_train.head()
```

## Building Model

Since we have a classification type problem, we can use Supervised or Unsupervised learning. As we have a specific output i.e Income, we'll be using Supervised learning methods like KNN, Naive Bayes, SVM, Decision Tree, Random Forest and select the best model.

#### Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score 
```

```python
# Building model using 5 nearest neighbours
knn_classifier = KNeighborsClassifier(n_neighbors=5)
knn_classifier.fit(X_train, y_train)
```

```python
# Predicting the Model
y_pred = knn_classifier.predict(X_test)
```

```python
# Checking accuracy
knn_accuracy = accuracy_score(y_test, y_pred)
knn_accuracy*100
```

```python
# Using 7 nearest neighbours
knn_classifier = KNeighborsClassifier(n_neighbors=7)  
knn_classifier.fit(X_train, y_train)
y_pred = knn_classifier.predict(X_test)
knn_accuracy = accuracy_score(y_test, y_pred)
knn_accuracy*100
```

```python
# Using 11 nearest neighbours
knn_classifier = KNeighborsClassifier(n_neighbors=11)  
knn_classifier.fit(X_train, y_train)
y_pred = knn_classifier.predict(X_test)
knn_accuracy = accuracy_score(y_test, y_pred)
knn_accuracy*100
```

```python
# Using 15 nearest neighbours
knn_classifier = KNeighborsClassifier(n_neighbors=15)  
knn_classifier.fit(X_train, y_train)
y_pred = knn_classifier.predict(X_test)
knn_accuracy = accuracy_score(y_test, y_pred)
knn_accuracy*100
```

```python
# Using 17 nearest neighbours
knn_classifier = KNeighborsClassifier(n_neighbors=17)
knn_classifier.fit(X_train, y_train)
y_pred = knn_classifier.predict(X_test)
knn_accuracy = accuracy_score(y_test, y_pred)
knn_accuracy*100
```

```python
# Using 19 nearest neighbours
knn_classifier = KNeighborsClassifier(n_neighbors=19)  
knn_classifier.fit(X_train, y_train)
y_pred = knn_classifier.predict(X_test)
knn_accuracy = accuracy_score(y_test, y_pred)
knn_accuracy*100
```

Since there was not much increase in accuracy, a 17 neighbour KNN model is the best KNN model for us.
17-KNN has an accuracy of 79.8%

Since our Random Forest model has the best accuracy out of KNN, NB, DT and RF, we'll use this as our final model.
